# Numerical condition analysis, asymmetric state probabilities

```python
# Imports, magics etc.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fontsize=12
linewidth=3
```

```python
# Read excel file
df = pd.read_excel('numanalysis.xlsx', skiprows=1)
```

## Learning equilibrium

```python
# Subsetting Dataframe for plot
df_learn = df[['alpha', 'U_learn_learn', 'U_learn_herd', 'U_learn_antiherd']]
df_learn.set_index('alpha')

fig, ax = plt.subplots(1,1, figsize=(10,7))

sns.lineplot(data=df_learn, x='alpha', y='U_learn_learn', ax=ax, color='red', alpha=0.5, label='Learning (eq. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_learn, x='alpha', y='U_learn_herd', ax=ax, color='blue', alpha=0.5, label='Herding (dev. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_learn, x='alpha', y='U_learn_antiherd', ax=ax, color='green', alpha=0.5,
             label='Anti-herding (dev. strat.)', linewidth=linewidth);

plt.axvline(0.5, linestyle='--', alpha=0.8, color='gray')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5, frameon=False, fontsize=fontsize);
ax.set_xlabel("$\\alpha$", fontsize=fontsize);
ax.set_ylabel("Expected utility", fontsize=fontsize);

plt.savefig('Figures/LearningEquilibriumUtility', dpi=300, bbox_inches='tight')
```

## Herding equilibrium

```python
# Subsetting Dataframe for plot
df_herd = df[['alpha', 'U_herd_herd', 'U_herd_learn', 'U_herd_antiherd']]
df_herd.set_index('alpha')

fig, ax = plt.subplots(1,1, figsize=(10,7))

sns.lineplot(data=df_herd, x='alpha', y='U_herd_learn', ax=ax, color='red', alpha=0.5, label='Learning (dev. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_herd, x='alpha', y='U_herd_herd', ax=ax, color='blue', alpha=0.5, label='Herding (eq. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_herd, x='alpha', y='U_herd_antiherd', ax=ax, color='green', alpha=0.5,
             label='Anti-herding (dev. strat.)', linewidth=linewidth);

plt.axvline(0.5, linestyle='--', alpha=0.8, color='gray')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5, frameon=False, fontsize=fontsize);
ax.set_xlabel("$\\alpha$", fontsize=fontsize);
ax.set_ylabel("Expected utility", fontsize=fontsize);

plt.savefig('Figures/HerdingEquilibriumUtility', dpi=300, bbox_inches='tight')
```

## Anti-herding equilibrium

```python
# Subsetting Dataframe for plot
df_antiherd = df[['alpha', 'U_antiherd_herd', 'U_antiherd_learn', 'U_antiherd_antiherd']]
df_antiherd.set_index('alpha')

fig, ax = plt.subplots(1,1, figsize=(10,7))

sns.lineplot(data=df_antiherd, x='alpha', y='U_antiherd_learn', ax=ax, color='red', alpha=0.5, label='Learning (dev. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_antiherd, x='alpha', y='U_antiherd_herd', ax=ax, color='blue', alpha=0.5, label='Herding (dev. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_antiherd, x='alpha', y='U_antiherd_antiherd', ax=ax, color='green', alpha=0.5,
             label='Anti-herding (eq. strat.)', linewidth=linewidth);

plt.axvline(0.5, linestyle='--', alpha=0.8, color='gray')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5, frameon=False, fontsize=fontsize);
ax.set_xlabel("$\\alpha$", fontsize=fontsize);
ax.set_ylabel("Expected utility", fontsize=fontsize);

plt.savefig('Figures/AntiherdingEquilibriumUtility', dpi=300, bbox_inches='tight')
```

## A's incentive to learn

```python
# Subsetting Dataframe for plot
df_A = df[['alpha', 'U_learn', 'U_randomize']]
df_A['U_learn_C'] = df_A['U_learn'] - 0.025
df_A.set_index('alpha')

fig, ax = plt.subplots(1,1, figsize=(10,7))

sns.lineplot(data=df_A, x='alpha', y='U_learn', ax=ax, color='red', alpha=0.6, label='Learning, $C=0$ (eq. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_A, x='alpha', y='U_learn_C', ax=ax, color='red', alpha=0.2, label='Learning, $C>0$ (eq. strat.)',
            linewidth=linewidth);

sns.lineplot(data=df_A, x='alpha', y='U_randomize', ax=ax, color='black', alpha=0.6,
             label='Randomize (dev. strat.)', linewidth=linewidth);

plt.axvline(0.5, linestyle='--', alpha=0.8, color='gray')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5, frameon=False, fontsize=fontsize);
ax.set_xlabel("$\\alpha$", fontsize=fontsize);
ax.set_ylabel("Expected utility", fontsize=fontsize);

plt.savefig('Figures/AUtilityLearnRandomize', dpi=300, bbox_inches='tight')
```

## THETA movements

```python
# Subsetting Dataframe for plot
df_theta = df[['alpha', 'thetaGGG', 'thetaGGB', 'thetaGBG', 'thetaGBB']]
df_theta.set_index('alpha')

fig, ax = plt.subplots(1,1, figsize=(10,7))

sns.lineplot(data=df_theta, x='alpha', y='thetaGGG', ax=ax, label='$\hat{\mu}^B(s_G,s_G,\\theta_G)$',
            linewidth=linewidth);
sns.lineplot(data=df_theta, x='alpha', y='thetaGGB', ax=ax, label='$\hat{\mu}^B(s_G,s_G,\\theta_B)$',
            linewidth=linewidth);
sns.lineplot(data=df_theta, x='alpha', y='thetaGBG', ax=ax, label='$\hat{\mu}^B(s_G,s_B,\\theta_G)$',
            linewidth=linewidth);
sns.lineplot(data=df_theta, x='alpha', y='thetaGBB', ax=ax, label='$\hat{\mu}^B(s_G,s_B,\\theta_B)$',
            linewidth=linewidth);

plt.axvline(0.5, linestyle='--', alpha=0.8, color='gray')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5, frameon=False, fontsize=fontsize);
ax.set_xlabel("$\\alpha$", fontsize=fontsize);
ax.set_ylabel("", fontsize=fontsize);

plt.savefig('Figures/ThetasAsymmetric', dpi=300, bbox_inches='tight')
```
